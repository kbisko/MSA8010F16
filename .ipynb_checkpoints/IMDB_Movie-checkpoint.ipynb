```python
all_features = set(movie.columns)-set(['imdb_score'])
num_features = set(movie.describe().columns)-set(['imdb_score'])-set(['Unnamed'])
cat_features = all_features-num_features

print("All features:         ", ", ".join(all_features), "\nNumerical features:   ", ", ".join(num_features), "\nCategorical features: ", ", ".join(cat_features))
```

```python
set(movie.columns)-set(movie.describe().columns)-set('imdb_score')
```

```python
movie['content_rating'].fillna('Unknown',inplace=True)
movie['aspect_ratio'].replace(16.0,1.77,inplace=True)
movie
```

```python
movie=movie[~movie['content_rating'].str.contains('TV')]
movie
```
