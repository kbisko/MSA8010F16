```python
gen=movie.loc[:,["movie_title","genres"]].groupby('movie_title').genres.apply(lambda x: pd.DataFrame(x.str.split('|').tolist())).reset_index().drop('level_1', axis=1)
gen
```

```python
genre_list=movie['genres'].apply(lambda x: x.split('|'))
flat_gen_list=set()
for x in genre_list:
    for y in x:
        flat_gen_list.add(y)
flat_gen_list 
```

```python
genre_list_final=[]
for y in range(0,len(gen)):
    for x in genre_list:
        for i in range(1,8):
            if gen.iloc[y,i] == x: 
                z = 1
                break
            else:
                z = 0
                i+1
        genre_list_final.append([gen.loc[y,'movie_title'], x, z])

genre_list_final
```
