# Part 1 - Observations from the data


### 1. Do larcenies follow the same seasonal trends? Or does it very per type?

```python
larcv = resdf['LARCENY-NON VEHICLE']
larcnv = resdf['LARCENY-FROM VEHICLE']
```

```python
%matplotlib inline
fig = plt.figure(figsize=(10,5))
larcv1=np.transpose(larcv)
larcnv1=np.transpose(larcnv)
#plt.plot(larc1, )
plt.plot(larcv1, color='blue')
plt.plot(larcnv1, color='green')
plt.title('LARCENY')
plt.xticks(range(13), ['', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
fig.savefig('Larceny_over_month.svg')
'done'
```

#### The blue line, which is non-vehicular larceny peaks in the fall and takes a dramatic drop in December. This seems very strange as general theft stats suggest that the Summer and December are the most prevelant times. The green line, vehicular larceny does follow this trend with a peak in August and another in December. The two lines both have a sharp drop in February, which is to be expected since it has less days then other months, but it also follows the trend of a drop after the holiday season. 

### 2. What are the top 10 neighborhoods for total offenses? Are there any that are dramatically worse?

```python
Neigh = df.groupby(['neighborhood']).offense_id.count().reset_index()
Neigh1 = Neigh.sort_values(by=['offense_id'], ascending=(False)).reset_index(drop=True)
Neigh2 = Neigh1.head(10).rename(columns={'offense_id':'Count'}).reset_index(drop=True)
Neigh2
```

```python
%matplotlib inline
width = 0.5       # the width of the bars
N = 10
ind = np.arange(N)  # the x locations for the groups

fig, ax = plt.subplots()
rects1 = ax.bar(ind, Neigh2.Count, width, color='r')

#plt.ylim(0, 500)
plt.title('Top 10 Neighborhoods for Crime')
plt.xticks(range(10), Neigh2.neighborhood, rotation=90)
# add some text for labels, title and axes ticks
#ax.set_xticklabels(autodf['AUTO THEFT'].index)

#ax.legend((rects1[0], rects2[0]), ('LARCENY', 'THEFT'))

fig.savefig('TopNeighborhoods.svg')
#x = 1
'done'
```

#### Downtown has significantly more crime than anywhere else in Atlanta. Midtown is the second highest with also dramtically more crime than else where. This could be due to their proximity to each other, along with their land size and population. These are two of the largest areas in Atlanta so it makes sense that they would have more crime. 

### 3. Do larcenies occur more often at certain times of day?

```python
df['occur_hour'] = df['occur_ts'].map(lambda x: x.hour)
time = df.groupby(['UC2 Literal', 'occur_hour']).offense_id.count().reset_index().rename(columns={'UC2 Literal':'Offense', 'offense_id':'Count'})

timev = time.loc[time.Offense=='LARCENY-NON VEHICLE', ['occur_hour', 'Count']].rename(columns={'Count':'Non_Vehicle'})
timenv = time.loc[time.Offense=='LARCENY-FROM VEHICLE', ['occur_hour', 'Count']].rename(columns={'Count':'Vehicle'})
```

```python
%matplotlib inline
fig = plt.figure(figsize=(10,5))
plt.plot(timev.occur_hour, timev.Non_Vehicle, color='blue') 
plt.plot(timenv.occur_hour, timenv.Vehicle, color='green')
plt.title('LARCENY by Time of Day')
plt.xticks(range(25))
plt.legend(loc='upper left')
fig.savefig('Larceny_time_day.svg')
'done'
```

#### There is definitley a time trend for both types of larcecny, with the early hours of the morning being the low point of each. For non-vehicular larceny the high point is during the afternoon hours peaking at 12 and continuing steady to around 6. Vehicular larcens occur mostly in the later hours starting to rapidly increase around 4 in the afternoon, peaking around 7, and then slowly declining into the morning. The non-vehicular have peaks at 8, 12, and then throughout the afternoon. This is probably centered around when people are coming and going to work. The vehicular is pretty consistently increase and decreasing from 7pm. There is a small peak at noon before dropping back to the steady state. 

# Part 2 - Seasonal Model

```python
## load complete dataset
dff = pd.read_excel('/home/data/APD/COBRA083016.xlsx', sheetname='Query')
```

```python
dff.shape
```

```python
for evt in ['occur', 'poss']:
    dff['%s_ts'%evt] = pd.to_datetime(dff['%s_date'%evt]+' '+dff['%s_time'%evt])
dff['rpt_ts'] = pd.to_datetime(dff.rpt_date)
```

```python
', '.join(dff.columns)
```

```python
dff['occur_year'] = dff.occur_ts.dt.year
dff['occur_month'] = dff.occur_ts.dt.month
dff['occur_dayweek'] = dff.occur_ts.dt.dayofweek
```

### Crime per year
Let's look at the 

```python
crime_year = dff[dff.occur_year.between(2009, 2015)].groupby(by=['UC2 Literal', 'occur_year']).offense_id.count()
```

```python
%matplotlib inline
fig = plt.figure(figsize=(40,30))
crime_types = crime_year.index.levels[0]
years = crime_year.index.levels[1]
for c in range(len(crime_types)):
    y_max = max(crime_year.loc[crime_types[c]])
    
    plt.subplot(4,3,c+1)
    plt.hlines(crime_year.loc[crime_types[c]].iloc[-1]*100/y_max, years[0], years[-1], linestyles="dashed", color="r")
    plt.bar(crime_year.loc[crime_types[c]].index, crime_year.loc[crime_types[c]]*100/y_max, label=crime_types[c], alpha=0.5)
    ##plt.legend()
    plt.ylim(0, 100)
    plt.xticks(years+0.4, [str(int(y)) for y in years], rotation=0, fontsize=24)
    plt.yticks([0,20,40,60,80,100], ['0%','20%','40%','60%','80%','100%'], fontsize=24)
    plt.title(crime_types[c], fontsize=30)
    None
```

```python
crime_year_month = dff[dff.occur_year.between(2009, 2015)].groupby(by=['UC2 Literal', 'occur_year', 'occur_month']).offense_id.count()
```

Let's look at residential burglary.

```python
c = 3
crime_types[c]
```

```python
c = 3 ## 'BURGLARY-RESIDENCE'
resburglaries = crime_year_month.loc[crime_types[c]]
fig = plt.figure(figsize=(20,10))
for y in years:
    plt.plot(resburglaries.loc[y].index, resburglaries.loc[y], label=("%4.0f"%y))
plt.legend()
plt.title("Seasonal Trends - %s"%crime_types[c], fontsize=20)
plt.xticks(range(13), ['', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.xlim(0,13)
None
```

Normalized over the annual average

```python
c = 3 ## 'BURGLARY-RESIDENCE'
fig = plt.figure(figsize=(20,10))
for y in years:
    avg = resburglaries.loc[y].mean()
    plt.hlines(avg, 1, 13, linestyle='dashed')
    plt.plot(resburglaries.loc[y].index, resburglaries.loc[y], label=("%4.0f"%y))
plt.legend()
plt.title("Seasonal Trends - %s (with annuale averages)"%crime_types[c], fontsize=20)
plt.xticks(list(range(1,13)), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.xlim(0,13)
None
```

```python
c = 3 ## 'BURGLARY-RESIDENCE'
fig = plt.figure(figsize=(20,10))
for y in years:
    avg = resburglaries.loc[y].mean()
    std = resburglaries.loc[y].std()
    ##plt.hlines(avg, 1, 13, linestyle='dashed')
    plt.plot(resburglaries.loc[y].index, (resburglaries.loc[y]-avg)/std, label=("%4.0f"%y))
plt.legend()
plt.title("Seasonal Trends - %s (normalized)"%crime_types[c], fontsize=20)
plt.xticks(list(range(1,13)), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.xlim(0,13)
plt.ylabel("Standard deviations $\sigma_y$")
None
```
