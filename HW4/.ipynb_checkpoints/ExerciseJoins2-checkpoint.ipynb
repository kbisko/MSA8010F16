```python
import pandas as pd
import numpy as np
```

```python
Employees = pd.read_excel('/home/data/AdventureWorks/Employees.xls')
Territory = pd.read_excel('/home/data/AdventureWorks/SalesTerritory.xls')
Customers = pd.read_excel('/home/data/AdventureWorks/Customers.xls')
Orders = pd.read_excel('/home/data/AdventureWorks/ItemsOrdered.xls')
```

### 1a. Provide a count of the number of salespeople in the company

```python
Employees.merge(Territory, how='inner').EmployeeID.count()
```

### 1b. Instead of a count, list the salespeople along with their territories
Show columns: EmployeeID, FirstName, LastName, TerritoryName, CountryCode, Region

```python
salesterr=pd.merge(Employees,Territory, on='TerritoryID', how='inner')
salesterr[['EmployeeID', 'FirstName', 'LastName', 'Name', 'CountryCode', 'Region']]
```

### 1c. Now include in the result set all other non-sales employees

```python
salesterr=pd.merge(Employees,Territory, on='TerritoryID', how='left')
salesterr[['EmployeeID', 'FirstName', 'LastName', 'Name', 'CountryCode', 'Region']]
```

### 1d. Sort the results with the salespeople at the top

```python
salesterr=pd.merge(Employees,Territory, on='TerritoryID', how='left')
salesterr[['EmployeeID', 'FirstName', 'LastName', 'Name', 'CountryCode', 'Region']].sort_values('Region')
```

### 1e. It turns out we don't need territory information for our American salespeople
Only show information for territories outside the U.S.

```python
salesterr=pd.merge(Employees,Territory[Territory.CountryCode != 'US'], on='TerritoryID', how='left')
salesterr[['EmployeeID', 'FirstName', 'LastName', 'Name', 'CountryCode', 'Region']].sort_values('Region')
```

### 2a. Take the query from #1b above, but add information for each employee's manager
Show additional columns: ManagerFirstName, ManagerLastName, ManagerJobTitle

```python
salesteam=pd.merge(Employees,Territory, on='TerritoryID', how='inner')
mngr=pd.merge(salesteam,Employees, left_on='ManagerID',right_on='EmployeeID', how='left')
mngr[['EmployeeID_x', 'FirstName_x', 'LastName_x', 'Name','CountryCode', 'Region','FirstName_y', 'LastName_y', 'JobTitle_y']]
```

### 2b. Only show salespeople who report to Amy Alberts

```python
salesteam=pd.merge(Employees,Territory, on='TerritoryID', how='inner')
mngr=pd.merge(salesteam,Employees[Employees.FirstName.str.cat(Employees.LastName, sep='')=='AmyAlberts'], left_on='ManagerID',right_on='EmployeeID', how='left')
mngr[['EmployeeID_x', 'FirstName_x', 'LastName_x', 'Name','CountryCode', 'Region','FirstName_y', 'LastName_y', 'JobTitle_y']]
```

### 3a. Give a list of all customers, and the average price of items ordered for each one
Show columns: CustomerID, FirstName, LastName, City, StateName, AvgPrice

```python
corders = pd.merge(Customers[['CustomerID', 'FirstName', 'LastName', 'City', 'StateName']], Orders[['CustomerID', 'Price']], on='CustomerID', how='left')
corders.groupby(['CustomerID', 'FirstName', 'LastName', 'City', 'StateName']).Price.mean().reset_index()
```

### 3b. For those customers who have never ordered anything, put 0 as the average price

```python
corders = pd.merge(Customers[['CustomerID', 'FirstName', 'LastName', 'City', 'StateName']], Orders[['CustomerID', 'Price']], on='CustomerID', how='left')
corders.groupby(['CustomerID', 'FirstName', 'LastName', 'City', 'StateName']).Price.mean().reset_index().fillna(0)
```

### 3c. Show the same results, but instead of average price per customer, tell me the average "Total Price" per customer
Make sure you change the alias of the AvgPrice column accordingly

```python
corders = pd.merge(Customers[['CustomerID', 'FirstName', 'LastName', 'City', 'StateName']], Orders[['CustomerID', 'Price','Quantity']], on='CustomerID', how='left')
corders['Total']=corders.Price*corders.Quantity
corders.groupby(['CustomerID', 'FirstName', 'LastName', 'City', 'StateName']).Total.mean().reset_index().fillna(0)
```

### 4a. Someone has requested a report of all employees and their managers (if they don't have a manager, still display them)
Show columns: EmployeeID, EmployeeFullName, EmployeeTitle, BossFullName, BossTitle

### 9a. We need a list of each employee's boss's boss (i.e., 2 levels up)
Columns: EmployeeID, EmployeeFullName, EmployeeTitle, BossBossFullName, BossBossTitle<br/>
Sort by BossBossTitle and then EmployeeTitle

### 10a. We need a list of each manager's indirect reports (i.e., 2 levels down)
Columns: ManagerFullName, ManagerTitle, IndirectReportFullName, IndirectReportTitle<br />
Sort by ManagerTitle and IndirectReportTitle

```python
Employees['FullName']=Employees.FirstName.str.cat(Employees.LastName, sep=' ')
emp_emp = pd.merge(pd.merge(Employees[['EmployeeID','FullName','JobTitle']], Employees[['EmployeeID', 'ManagerID']], left_on='EmployeeID', right_on='ManagerID', how='inner'), Employees[['ManagerID', 'EmployeeID','FullName','JobTitle' ]], left_on='EmployeeID_y', right_on='ManagerID', how='inner')
emp_emp[['FullName_x', 'JobTitle_x', 'FullName_y', 'JobTitle_y']].rename(columns=({'FullName_x':'ManagerFullName', 'JobTitle_x':'ManagerTitle', 'FullName_y':'IndirectFullName', 'JobTitle_y':'IndirectTitle'})).sort_values('ManagerTitle')
```

### 10b. Please tell us how many indirect reports each manager has
Columns: ManagerFullName, ManagerTitle, NumIndirectReports

```python
Employees['FullName']=Employees.FirstName.str.cat(Employees.LastName, sep=' ')
emp_emp = pd.merge(pd.merge(Employees[['EmployeeID','FullName','JobTitle']], Employees[['EmployeeID', 'ManagerID']], left_on='EmployeeID', right_on='ManagerID', how='inner'), Employees[['ManagerID']], left_on='EmployeeID_y', right_on='ManagerID', how='inner')
emp_emp.groupby(['FullName', 'JobTitle']).ManagerID_y.count().reset_index().rename(columns=({'FullName':'ManagerFullName', 'JobTitle':'ManagerTitle', 'ManagerID_y':'MunIndirectReports'})).sort_values('ManagerTitle')
```
