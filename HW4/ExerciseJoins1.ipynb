### 4e. For the Arizona customers above, show the maximum total price per customer. Choose the columns wisely

```python
orders=Customers[Customers.StateName=='Arizona'].merge(Orders, how='left', on='CustomerID')
orders['Total']=orders.Quantity*orders.Price
orders.groupby([orders.CustomerID, orders.FirstName.str.cat(orders.LastName, sep=' ')]).Total.max()
```

### 4f. For the results above, only show the ones where the MaxTotalPrice is more than $5.00

```python
regions = Territory.merge(Customers, right_on='SalesTerritoryID', left_on='TerritoryID', how='left')
regions['CityRegion']= regions.City.str.cat(regions.Region, sep=' - ')
regions
```

### BONUS 2: Fix the problem above where CityRegion is NULL sometimes; change it to, for example, "n/a - North America"

```python
regions = Territory.merge(Customers, right_on='SalesTerritoryID', left_on='TerritoryID', how='left')
regions.fillna('n/a', inplace=True)
regions['CityRegion']= regions.City.str.cat(regions.Region, sep=' - ')
regions
```
