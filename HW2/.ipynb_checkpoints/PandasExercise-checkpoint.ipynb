# Filtering (with)

### 1. Provide a list of employees that are married.

```python
Married = Employees[Employees['MartialStatus'].isin(['M'])]
Married.loc[:,['EmployeeID','FirstName','LastName']]
```

### 2a. Show me a list of employees that have a lastname that begins with "R".

```python
Rnames = Employees[Employees.LastName.str.contains('R.*')]
Rnames.loc[:,['EmployeeID','FirstName','LastName']]
```

### 2b. Show me a list of employees that have a lastname that ends with "r"

```python
namesr = Employees[Employees.LastName.str.contains('.*r$')]
namesr.loc[:,['EmployeeID','FirstName','LastName']]
```

### 2c. Provide a list of employees that have a hyphenated lastname.

```python
Hynames = Employees[Employees.LastName.str.contains('.*-.*')]
Hynames.loc[:,['EmployeeID','FirstName','LastName']]
```

### 3a. Provide a list of employees that are on salary and have more than 35 vacation hours left.

```python
vaca1 = Employees[(Employees.SalariedFlag==1) & (Employees.VacationHours>35)]
vaca1.loc[:,['EmployeeID','FirstName','LastName','VacationHours','SalariedFlag']]
```

### 3b. Show the same as above but limit it to American employees.

```python
vaca2 = Employees[(Employees.SalariedFlag==1) & (Employees.VacationHours>35) & (Employees.CountryName == 'United States')]
vaca2.loc[:,['EmployeeID','FirstName','LastName','VacationHours','SalariedFlag','CountryName']]
```

### 3c. Show the same as above but limit it to non-American employees.

```python
vaca3 = Employees[(Employees.SalariedFlag==1) & (Employees.VacationHours>35) & (Employees.CountryName != 'United States')]
vaca3.loc[:,['EmployeeID','FirstName','LastName','VacationHours','SalariedFlag','CountryName']]
```

### 4a. List the married employees with more than 35 vacation hours, only ones living in Washington state.

```python
vaca4 = Employees[(Employees.MaritalStatus=='M') & (Employees.VacationHours>35) & (Employees.StateProvinceName == 'Washington')]
vaca4.loc[:,['EmployeeID','FirstName','LastName','MaritalStatus','VacationHours','SalariedFlag','StateProvinceName','CountryName']]
```

### 4b. Change the logic to include anyone who meets any of the 3 conditions (i.e., people who are either married, live in Washington state, or have more than 35 vacation hours left)

```python
vaca5 = Employees[(Employees.MaritalStatus=='M') | (Employees.VacationHours>35) | (Employees.StateProvinceName == 'Washington')]
vaca5.loc[:,['EmployeeID','FirstName','LastName','MaritalStatus','VacationHours','SalariedFlag','StateProvinceName','CountryName']]
```

### 4c. Show the same as above, but only for Production Technicians

```python
vaca6 = Employees[(Employees.JobTitle.str.contains('Production Technician.*')) & ((Employees.MaritalStatus=='M') | (Employees.VacationHours>35) | (Employees.StateProvinceName == 'Washington'))]
vaca6.loc[:,['EmployeeID','FirstName','LastName','MaritalStatus','JobTitle','VacationHours','SalariedFlag','StateProvinceName','CountryName']]
```

### 5a. List all employees living in Redmond, Seattle, and Bellevue, showing EmployeeID, FirstName, LastName, and City. Sort the list alphabetically by city.

```python
city1 = Employees[(Employees.City=='Redmond') | (Employees.City=='Seattle') | (Employees.City=='Bellevue')]
city12 = city1.loc[:,['EmployeeID','FirstName','LastName','City']].sort_values(by='City')
city12
```

### 5b. For the list above, make sure these are only in the state of Washington, just to be careful.

```python
city2 = Employees[(Employees.StateProvinceName=='Washington')&((Employees.City=='Redmond') | (Employees.City=='Seattle') | (Employees.City=='Bellevue'))]
city22 = city2.loc[:,['EmployeeID','FirstName','LastName','City','StateProvinceName']].drop_duplicates().sort_values(by='City')
city22
```

### 6. Provide a list of employees who have no title, whether it's a NULL or empty string.

```python
notitle = Employees[(Employees.Title.isnull()) | (Employees.Title == '')]
notitle.loc[:,['EmployeeID','Title','FirstName','LastName']]
```

### 7a. Provide a list of employees who have at least 60 vacation hours left.

```python
vacemp = Employees[(Employees.VacationHours>=60)]
vacemp.loc[:,['EmployeeID','Title','FirstName','LastName','VacationHours','SickLeaveHours']]
```

### 7b. Provide a list of employees who have less than 60 vacation hours left.

```python
SELECT e.EmployeeID, e.Title, e.FirstName, e.LastName, e.VacationHours, e.SickLeaveHours
	FROM dbo.Employees AS e
	WHERE e.VacationHours < 60
	;
```

```python
vacemp2 = Employees[(Employees.VacationHours<60)]
vacemp2.loc[:,['EmployeeID','Title','FirstName','LastName','VacationHours','SickLeaveHours']]
```

### 7c. Show me employees who have more than 20 and less than 60 vacation hours left.

```python
vacemp = Employees[(Employees.VacationHours<60)&(Employees.VacationHours>20)]
vacemp.loc[:,['EmployeeID','Title','FirstName','LastName','VacationHours','SickLeaveHours']].sort_values(by='VacationHours')
```

### 7d. If you did not use BETWEEN for 7c, do the same but use BETWEEN. If you did use BETWEEN for 7c, do it another way.

# Grouping

### 1a. What is the earliest birthdate for all employees?

```python
Employees.BirthDate.min()
```

### 1b. Add to the above, the most recent birthdate for all employees

```python
Employees.BirthDate.min(), Employees.BirthDate.max()
```

### 1c. Show the above results broken down by gender

```python
Employees.groupby(Employees.Gender,sort=False, as_index=False).agg({'BirthDate': [np.min, np.max]})
```

### 1d. Show the above results broken down by gender, and salaried/hourly

```python
Employees.groupby([Employees.Gender, Employees.SalariedFlag],sort=False, as_index=False).agg({'BirthDate': [np.min, np.max]})
```

### 2a. What are the average vacation hours for all employees?

```python
Employees.VacationHours.mean()
```

### 2b. Add to the above, the minimum vacation hours for all employees

```python
Employees.VacationHours.mean(), Employees.VacationHours.min()
```

### 2c. Show the above results broken down and ordered by job title

```python
Employees.groupby(Employees.JobTitle,sort=True, as_index=False).agg({'VacationHours': [ np.mean, np.min]})
```

### 2d. Show the above results broken down by job title, and married/single employees

```python
Employees.groupby([Employees.JobTitle, Employees.MaritalStatus],sort=True, as_index=False).agg({'VacationHours': [ np.mean, np.min]})
```

### 2e. Add to the above, the maximum vacation hours per group

```python
Employees.groupby([Employees.JobTitle, Employees.MaritalStatus],sort=True, as_index=False).agg({'VacationHours': [ np.mean, np.min, np.max]})
```

### 2f. Show the above results broken down by job title, married/single employees, and State

```python
Employees.groupby([Employees.JobTitle, Employees.MaritalStatus, Employees.StateProvinceName],sort=True, as_index=False).agg({'VacationHours': [np.mean, np.min, np.max]})
```

### 2g. Show the above results but only for American employees

```python
Employees[(Employees.CountryName=='United States')].groupby([Employees.JobTitle, Employees.MaritalStatus, Employees.StateProvinceName]).agg({'VacationHours': [np.mean, np.min, np.max]})
```

### 2h. Change the grouping above so it's broken down by married/single and State, no more job title

```python
Employees[(Employees.CountryName=='United States')].groupby([Employees.MaritalStatus, Employees.StateProvinceName]).agg({'VacationHours': [np.mean, np.min, np.max]})
```

### 2i. Limit the results above to States where the average vacation hours is greater than 30

```python
mar_state_emp = Employees[(Employees.CountryName=='United States')]
mar_state_emp2 = mar_state_emp.groupby([Employees.MaritalStatus, Employees.StateProvinceName]).filter(lambda x: x['VacationHours'].mean() > 30.0)
mar_state_emp2.groupby([Employees.MaritalStatus, Employees.StateProvinceName]).agg({'VacationHours': [np.mean,np.min,np.max]})
```

```python
state_emp = Employees[(Employees.CountryName=='United States')]
state_emp2 = state_emp.groupby([Employees.StateProvinceName]).filter(lambda x: x['VacationHours'].mean() > 30.0)
state_emp2.groupby([Employees.StateProvinceName]).agg({'VacationHours': [np.mean,np.min,np.max]})
```

### 2j. Limit the results above to States where the average vacation hours is greater than 30 and the maximum vacation hours is less than 50

```python
state_emp = Employees[(Employees.CountryName=='United States')]
state_emp2 = state_emp.groupby([Employees.StateProvinceName]).filter(lambda x: 30.0 < x['VacationHours'].mean() < 49.0)
state_emp2.groupby([Employees.StateProvinceName]).agg({'VacationHours': [np.mean,np.min,np.max]})
```

### 2k. Show the same results but only for non-American employees

```python
state_emp = Employees[(Employees.CountryName!='United States')]
state_emp2 = state_emp.groupby([Employees.StateProvinceName]).filter(lambda x: 30.0 < x['VacationHours'].mean() < 49.0)
state_emp2.groupby([Employees.StateProvinceName]).agg({'VacationHours': [np.mean,np.min,np.max]})
```

### 3a. Report how many employees are in the company

```python
Employees['EmployeeID'].count()
```

### 3b. For the above report, show the number of employees per manager (hint: use ManagerID)

```python
emp_cnt = Employees.groupby(Employees.ManagerID,sort=False, as_index=False).agg({'EmployeeID': 'count'})
emp_cnt.rename(columns={'EmployeeID': 'NumEmployees'})
```

### 3c. Remove any manager ID's that are NULL from the results above

```python
emp_cnt_na = Employees.dropna(subset=['ManagerID'])
emp_cnt_na2 = emp_cnt_na.groupby(emp_cnt_na.ManagerID,sort=False, as_index=False).agg({'EmployeeID': 'count'})
emp_cnt_na2.rename(columns={'EmployeeID': 'NumEmployees'})
```

### 3d. Show the same results as above, but only for managers who have at least 5 employees

```python
manage = Employees.dropna(subset=['ManagerID'])
manage2 = manage.groupby(manage.ManagerID,sort=False, as_index=False).filter(lambda x: x['EmployeeID'].count()>5)
manage3 = manage2.groupby(manage2.ManagerID,sort=False, as_index=False).agg({'EmployeeID': 'count'})
manage3.rename(columns={'EmployeeID': 'NumEmployees'})
```

```python
Employees['VacationHours'].mean()
```

### 4b. Break down the results by State

```python
vaca = Employees.groupby(Employees.StateProvinceName,sort=False, as_index=False).agg({'VacationHours': 'mean'})
vaca.rename(columns={'VacationHours': 'Average Vacation Hours'})
```

### 4c. Break down the results by city and State

```python
vaca2 = Employees.groupby([Employees.StateProvinceName,Employees.City], sort=False, as_index=False).agg({'VacationHours': 'mean'})
vaca2.rename(columns={'VacationHours': 'Average Vacation Hours'})
```

### 4d. Add something that shows the number of employees per city

```python
vaca3 = Employees.groupby([Employees.StateProvinceName,Employees.City], sort=False, as_index=False).agg({'VacationHours': 'mean', 'EmployeeID':'count'})
vaca3.rename(columns={'VacationHours': 'Average Vacation Hours', 'EmployeeID':'NumEmployees'})
```

### 4e. Sort the results by the city and state

```python
vaca4 = Employees.groupby([Employees.StateProvinceName,Employees.City], sort=False, as_index=False).agg({'VacationHours': 'mean', 'EmployeeID':'count'})
vaca4.sort_values(by=['StateProvinceName','City']).rename(columns={'VacationHours': 'Average Vacation Hours', 'EmployeeID':'NumEmployees'})
```

### 4f. Make city and State a single column in the format of "City, State"

```python
vaca5 = Employees.groupby([Employees.StateProvinceName,Employees.City], sort=False, as_index=False).agg({'VacationHours': 'mean', 'EmployeeID':'count'})
vaca52 = vaca5.sort_values(by=['StateProvinceName','City']).rename(columns={'VacationHours': 'Average Vacation Hours', 'EmployeeID':'NumEmployees'})
vaca52['Location'] = vaca52.City+', '+vaca52.StateProvinceName
vaca52[['Location','Average Vacation Hours','NumEmployees']]
```

### 4g. Add a column that shows the difference between the maximum vacation hours and minimum vacation hours for each city

```python
vaca5 = Employees.groupby([Employees.StateProvinceName,Employees.City], sort=False, as_index=False).agg({'VacationHours': ['mean', lambda x: max(x)-min(x)], 'EmployeeID':'count'})
vaca52 = vaca5.sort_values(by=['StateProvinceName','City']).rename(columns={'mean': 'Average Vacation Hours','<lambda>': 'Difference Vacation Hours', 'EmployeeID':'NumEmployees'})
vaca52['Location'] = vaca52.City+', '+vaca52.StateProvinceName
vaca52[[5,2,3,4]]
```

### 4h. Now sort the results by the new column created above

```python
vaca5 = Employees.groupby([Employees.StateProvinceName,Employees.City], sort=False, as_index=False).agg({'VacationHours': ['mean', lambda x: max(x)-min(x)], 'EmployeeID':'count'})
vaca52 = vaca5.sort_values(by=['StateProvinceName','City']).rename(columns={'mean': 'Average Vacation Hours','<lambda>': 'Difference Vacation Hours', 'EmployeeID':'NumEmployees'})
vaca53 = vaca52.sort_values(by=('VacationHours','Difference Vacation Hours'), ascending=False)
vaca53['Location'] = vaca53.City+', '+vaca53.StateProvinceName
vaca53[[5,2,3,4]]
```

### 4i. Limit the results to cities that have more than 1 employee

```python
vaca6 = Employees.groupby([Employees.StateProvinceName,Employees.City]).filter(lambda x: x['EmployeeID'].count() > 1 ) 
vaca62 = vaca6.groupby([Employees.City+', ' +Employees.StateProvinceName]).agg({'VacationHours': ['mean', lambda x: max(x)-min(x)], 'EmployeeID':'count'}).rename(columns={'mean': 'Average Vacation Hours','<lambda>': 'Difference Vacation Hours', 'EmployeeID':'NumEmployees'})
vaca63 = vaca62.sort_values(by=('VacationHours','Difference Vacation Hours'), ascending=False)
vaca63
```

### 4j. Limit the results to non-U.S. cities

```python
vaca7 = Employees[(Employees.CountryName!='United States')].groupby([Employees.StateProvinceName,Employees.City]).filter(lambda x: x['EmployeeID'].count() > 1 ) 
vaca72 = vaca7.groupby([Employees.City+', ' +Employees.StateProvinceName]).agg({'VacationHours': ['mean', lambda x: max(x)-min(x)], 'EmployeeID':'count'}).rename(columns={'mean': 'Average Vacation Hours','<lambda>': 'Difference Vacation Hours', 'EmployeeID':'NumEmployees'})
vaca73 = vaca72.sort_values(by=('VacationHours','Difference Vacation Hours'), ascending=False)
vaca73
```
